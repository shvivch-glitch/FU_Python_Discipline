# ЗАДАНИЕ 7. Сортировка списков.

```python
# №1
numbers = list(map(int, input("Введите числа через пробел: ").split()))

odd_numbers = [x for x in numbers if x % 2 != 0]
odd_numbers.sort(reverse=True)

print("Нечетные числа по убыванию:", odd_numbers)
```

```python
# №2
numbers = list(map(int, input("Введите числа через пробел: ").split()))

positive_numbers = [x for x in numbers if x > 0]
positive_numbers.sort()

print("Положительные числа по возрастанию:", positive_numbers)
```

```python
# №13
numbers = list(map(int, input("Введите числа через пробел: ").split()))

odd_positions = [numbers[i] for i in range(len(numbers)) if i % 2 != 0]
odd_positions.sort(reverse=True)

print("Числа на нечетных позициях по убыванию:", odd_positions)
```

```python
# №14
numbers = list(map(int, input("Введите числа через пробел: ").split()))

two_digit_numbers = [x for x in numbers if 10 <= abs(x) <= 99]
two_digit_numbers.sort()

print("Двузначные числа по возрастанию:", two_digit_numbers)
```

```python
# №15
numbers = list(map(int, input("Введите числа через пробел: ").split()))

even_positions = [numbers[i] for i in range(len(numbers)) if i % 2 == 0]
even_positions.sort()

print("Числа на четных позициях по возрастанию:", even_positions)
```
