# ЗАДАНИЕ 1. Ввод/вывод информации. Оператор =.

```python
# №1
import math
x = float(input("Введите x: "))
y = 3 * x**2 + math.sin(x + 2)
print(f"y = {y}")
```

```python
# №2
import math
a = float(input("Введите a: "))
x = float(input("Введите x: "))
y = a * x**2 + math.cos(2*x + 1)
print(f"y = {y}")
```

```python
# №3
import math
a = float(input("Введите a: "))
b = float(input("Введите b: "))
x = float(input("Введите x: "))
y = a * x + b * math.sin(2*x + 2)
print(f"y = {y}")
```
