```python
# №1
def Fact(N):
    if N == 0 or N == 1:
        return 1
    return N * Fact(N-1)

def Fact2(N):
    if N == 1 or N == 2:
        return N
    return N * Fact2(N-2)

print()
```

```python
# №2
def PowerN(x, n):
    if n == 0:
        return 1
    elif n > 0:
        return x * PowerN(x, n-1)
    else:
        return 1 / PowerN(x, -n)

print()
```

```python
# №3
def SqrtK(x, k, n):
    if n == 0:
        return 1
    y_prev = SqrtK(x, k, n-1)
    return y_prev - (y_prev - x / (y_prev ** (k-1))) / k

print()
```

```python
# №4
call_count = 0

def FibRec(N):
    global call_count
    call_count += 1
    if N == 1 or N == 2:
        return 1
    return FibRec(N-2) + FibRec(N-1)

def find_fibonacci_numbers():
    global call_count
    numbers = [5, 7, 10, 12, 15]
    results = []
    for num in numbers:
        call_count = 0
        fib_result = FibRec(num)
        results.append((num, fib_result, call_count))
    return results

print()
```

```python
# №5
call_count_C = 0

def C(m, n):
    global call_count_C
    call_count_C += 1
    if m == 0 or m == n:
        return 1
    return C(m, n-1) + C(m-1, n-1)

def find_combinations(N, M_values):
    global call_count_C
    results = []
    for M in M_values:
        call_count_C = 0
        comb_result = C(M, N)
        results.append((M, comb_result, call_count_C))
    return results

print()
```

```python
# №6
def NOD(A, B):
    if A == 0:
        return B
    return NOD(B % A, A)

def find_gcd_pairs(A, B, C, D):
    return NOD(A, B), NOD(A, C), NOD(A, D)

print()
```

```python
# №7
def MinRec(A, N):
    if N == 1:
        return A[0]
    return min(A[N-1], MinRec(A, N-1))

def find_min_elements(arrays):
    results = []
    for arr in arrays:
        results.append(MinRec(arr, len(arr)))
    return results

print()
```

```python
# №8
def Digits(S):
    if len(S) == 0:
        return 0
    return (1 if S[0].isdigit() else 0) + Digits(S[1:])

def count_digits_in_strings(strings):
    results = []
    for s in strings:
        results.append(Digits(s))
    return results

print()
```

```python
# №9
def Simm(S):
    if len(S) <= 1:
        return True
    if S[0] != S[-1]:
        return False
    return Simm(S[1:-1])

def check_symmetric_strings(strings):
    results = []
    for s in strings:
        results.append(Simm(s))
    return results

print()
```

```python
# №10
def binary_addition(a, b):
    max_len = max(len(a), len(b))
    a = a.zfill(max_len)
    b = b.zfill(max_len)

    def add_bin(x, y, carry=0):
        if not x:
            return str(carry) if carry else ''
        bit_x = int(x[-1])
        bit_y = int(y[-1])
        total = bit_x + bit_y + carry
        return add_bin(x[:-1], y[:-1], total // 2) + str(total % 2)

    result = add_bin(a, b)
    return result if result else '0'

def handle_negative_binary(pos_bin, neg_bin):
    pos_num = int(pos_bin, 2)
    neg_num = -int(neg_bin, 2)
    result = pos_num + neg_num
    return bin(result & 0b11111111)[2:] if result < 0 else bin(result)[2:]

print()
```

```python
# №11
def Root(f, a, b, epsilon):
    mid = (a + b) / 2
    if abs(f(mid)) < epsilon or (b - a) < epsilon:
        return mid
    if f(a) * f(mid) < 0:
        return Root(f, a, mid, epsilon)
    else:
        return Root(f, mid, b, epsilon)

print()
```

```python
# №12
def reverse_number(n, reversed_n=0):
    if n == 0:
        return reversed_n
    return reverse_number(n // 10, reversed_n * 10 + n % 10)

print()
```

```python
# №13
def Simm_part(S, i, j):
    if i >= j:
        return True
    if S[i] != S[j]:
        return False
    return Simm_part(S, i+1, j-1)

def check_partial_symmetry(strings, indices):
    results = []
    for s, (i, j) in zip(strings, indices):
        results.append(Simm_part(s, i, j))
    return results

print()
```

```python
# №14
def sum_digits(N):
    if N == 0:
        return 0
    return N % 10 + sum_digits(N // 10)

print()
```

```python
# №15
def is_palindrome(word, start=0, end=None):
    if end is None:
        end = len(word) - 1
    if start >= end:
        return "yes"
    if word[start] != word[end]:
        return "no"
    return is_palindrome(word, start + 1, end - 1)

def check_palindromes(words):
    results = []
    for word in words:
        results.append(is_palindrome(word))
    return results

print()
```
