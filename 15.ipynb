```python
# №1
M = int(input("Введите M: "))
N = int(input("Введите N: "))
A = []
for i in range(M):
    row = list(map(int, input(f"Введите строку {i+1}: ").split()))
    A.append(row)
K = int(input("Введите K: "))
L = int(input("Введите L: "))

if K >= M or L >= N:
    print("Матрица не изменяется")
else:
    result = []
    for i in range(M):
        if i == K:
            continue
        row = []
        for j in range(N):
            if j == L:
                continue
            row.append(A[i][j])
        result.append(row)
    print("Результат:")
    for row in result:
        print(row)
```

```python
# №2
A = list(map(int, input("Введите список A: ").split()))
B = []
C = []
for num in A:
    if num % 2 == 0:
        B.append(num)
    else:
        C.append(num)
print(f"Четные: {B}")
print(f"Нечетные: {C}")
```

```python
# №3
M = int(input("Введите M: "))
N = int(input("Введите N: "))
A = []
for i in range(M):
    row = list(map(float, input(f"Введите строку {i+1}: ").split()))
    A.append(row)
K1 = int(input("Введите K1: "))
K2 = int(input("Введите K2: "))

if K1 >= M or K2 >= M:
    print("Неверные индексы строк")
else:
    A[K1], A[K2] = A[K2], A[K1]
    print("Результат:")
    for row in A:
        print(row)
```

```python
# №4
M = int(input("Введите порядок матрицы M: "))
A = []
for i in range(M):
    row = list(map(int, input(f"Введите строку {i+1}: ").split()))
    A.append(row)

result = [[0] * M for _ in range(M)]
for i in range(M):
    for j in range(M):
        result[j][i] = A[i][j]
print("Транспонированная матрица:")
for row in result:
    print(row)
```

```python
# №5
rows_A = int(input("Введите количество строк A: "))
cols_A = int(input("Введите количество столбцов A: "))
A = []
for i in range(rows_A):
    row = list(map(int, input(f"Введите строку {i+1} матрицы A: ").split()))
    A.append(row)

rows_B = int(input("Введите количество строк B: "))
cols_B = int(input("Введите количество столбцов B: "))
B = []
for i in range(rows_B):
    row = list(map(int, input(f"Введите строку {i+1} матрицы B: ").split()))
    B.append(row)

if cols_A != rows_B:
    print("Умножение невозможно")
else:
    result = [[0] * cols_B for _ in range(rows_A)]
    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                result[i][j] += A[i][k] * B[k][j]
    print("Результат умножения:")
    for row in result:
        print(row)
```

```python
# №6
rows = int(input("Введите количество строк: "))
cols = int(input("Введите количество столбцов: "))
A = []
for i in range(rows):
    row = list(map(int, input(f"Введите строку {i+1} матрицы A: ").split()))
    A.append(row)

B = []
for i in range(rows):
    row = list(map(int, input(f"Введите строку {i+1} матрицы B: ").split()))
    B.append(row)

result = [[0] * cols for _ in range(rows)]
for i in range(rows):
    for j in range(cols):
        result[i][j] = A[i][j] + B[i][j]
print("Сумма матриц:")
for row in result:
    print(row)
```

```python
# №7
n = int(input("Введите порядок матрицы: "))
A = []
for i in range(n):
    row = list(map(int, input(f"Введите строку {i+1}: ").split()))
    A.append(row)

choice = input("Выберите диагональ (главная/побочная): ")
if choice == "главная":
    total = 0
    for i in range(n):
        total += A[i][i]
    print(f"Сумма главной диагонали: {total}")
else:
    total = 0
    for i in range(n):
        total += A[i][n-1-i]
    print(f"Сумма побочной диагонали: {total}")
```

```python
# №8
M = int(input("Введите M: "))
N = int(input("Введите N: "))
A = []
for i in range(M):
    row = list(map(int, input(f"Введите строку {i+1}: ").split()))
    A.append(row)

K1 = int(input("Введите K1: "))
K2 = int(input("Введите K2: "))

if K1 >= M:
    print("Матрица не изменяется")
else:
    K2 = min(K2, M-1)
    result = []
    for i in range(M):
        if K1 <= i <= K2:
            continue
        result.append(A[i])
    print("Результат:")
    for row in result:
        print(row)
```

```python
# №9
S = input("Введите строку S: ")
K = int(input("Введите K: "))
N = int(input("Введите N: "))

if K >= len(S):
    result = ""
else:
    end = min(K + N, len(S))
    substring = S[K:end]
    result = substring[::-1]
print(f"Результат: {result}")
```

```python
# №11
S = input("Введите строку S: ")

if not S:
    result = ""
else:
    result = []
    count = 1
    current_char = S[0]
    for i in range(1, len(S)):
        if S[i] == current_char:
            count += 1
        else:
            if count > 4:
                result.append(f"{current_char}{{{count}}}")
            else:
                result.append(current_char * count)
            current_char = S[i]
            count = 1
    if count > 4:
        result.append(f"{current_char}{{{count}}}")
    else:
        result.append(current_char * count)
    result = "".join(result)
print(f"Сжатая строка: {result}")
```

```python
# №12
S = input("Введите строку S: ")

if not S or S[0].isdigit():
    is_valid = False
else:
    is_valid = True
    for char in S:
        if not (char.isalnum() or char == '_'):
            is_valid = False
            break
print(f"Допустимый идентификатор: {is_valid}")
```

```python
# №13
M = int(input("Введите M: "))
N = int(input("Введите N: "))

A = [[0] * N for _ in range(M)]
for i in range(M):
    for j in range(N):
        A[i][j] = (i + j) % 2
print("Шахматная доска:")
for row in A:
    print(row)
```

```python
# №14
L = list(map(int, input("Введите список L: ").split()))

if not L:
    result = []
else:
    sorted_L = sorted(L)
    result = []
    left = True
    while sorted_L:
        if left:
            result.append(sorted_L.pop(0))
        else:
            result.append(sorted_L.pop())
        left = not left
print(f"Результат: {result}")
```
