```python
# №1
filename = input()
with open(filename, 'r') as f:
    numbers = list(map(float, f.read().split()))
product = 1
for num in numbers:
    product *= num
print(product)
```

```python
# №2
filename = input()
with open(filename, 'r') as f:
    numbers = list(map(int, f.read().split()))
positive = [x for x in numbers if x > 0]
negative = [x for x in numbers if x < 0]
result = positive + negative
with open('g.txt', 'w') as g:
    g.write(' '.join(map(str, result)))
```

```python
# №3
filename = input()
with open(filename, 'r') as f:
    numbers = list(map(int, f.read().split()))
squares = [x for x in numbers if int(x**0.5)**2 == x]
with open('g.txt', 'w') as g:
    g.write(' '.join(map(str, squares)))
```

```python
# №5
filename = input()
with open(filename, 'r') as f:
    dates = f.readlines()
years = []
for date in dates:
    day, month, year = map(int, date.split())
    years.append(year)
print(min(years))
```

```python
# №6
filename = input()
with open(filename, 'r') as f:
    numbers = list(map(int, f.read().split()))
count_even = sum(1 for x in numbers if x % 2 == 0)
print(count_even)
```

```python
# №7
filename = input()
with open(filename, 'r') as f:
    first_two = f.read(2)
if first_two.isdigit():
    num = int(first_two)
    if num % 2 == 0:
        print(True)
    else:
        print(False)
else:
    print(False)
```

```python
# №10
filename = input()
with open(filename, 'r') as f:
    dates = f.readlines()
spring_dates = []
for date in dates:
    day, month, year = map(int, date.split())
    if 3 <= month <= 5:
        spring_dates.append(f"{day} {month} {year}")
for date in spring_dates:
    print(date)
```

```python
# №13
filename = input()
with open(filename, 'r') as f:
    numbers = list(map(int, f.read().split()))
even_numbers = [x for x in numbers if x % 2 == 0]
odd_numbers = [x for x in numbers if x % 2 != 0]
with open('g.txt', 'w') as g:
    g.write(' '.join(map(str, even_numbers)))
with open('h.txt', 'w') as h:
    h.write(' '.join(map(str, odd_numbers)))
```

```python
# №14
filename = input()
with open(filename, 'r') as f:
    numbers = list(map(int, f.read().split()))
positive = [x for x in numbers if x > 0]
negative = [x for x in numbers if x < 0]
result = []
for i in range(len(positive)):
    result.append(positive[i])
    result.append(negative[i])
with open('g.txt', 'w') as g:
    g.write(' '.join(map(str, result)))
```

```python
# №15
filename = input()
with open(filename, 'r') as f:
    numbers = list(map(int, f.read().split()))
positive = [x for x in numbers if x > 0]
negative = [x for x in numbers if x < 0]
result = []
for i in range(0, len(positive), 2):
    result.extend(positive[i:i+2])
    result.extend(negative[i:i+2])
with open('g.txt', 'w') as g:
    g.write(' '.join(map(str, result)))
```
