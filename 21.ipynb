```python
# №1
def read_matrices(filename):
    with open(filename, 'r') as f:
        data = f.read().strip().split('\n\n')
    matrices = []
    for matrix_str in data:
        rows = matrix_str.strip().split('\n')
        matrix = [list(map(int, row.split())) for row in rows]
        matrices.append(matrix)
    return matrices

def write_matrices(filename, matrices):
    with open(filename, 'w') as f:
        for matrix in matrices:
            for row in matrix:
                f.write(' '.join(map(str, row)) + '\n')
            f.write('\n')

file1 = input("Введите имя первого файла: ")
file2 = input("Введите имя второго файла: ")

matrices1 = read_matrices(file1)
matrices2 = read_matrices(file2)

to_move = []
remaining = []
for matrix in matrices1:
    if matrix[0][0] == 0:
        to_move.append(matrix)
    else:
        remaining.append(matrix)

write_matrices(file1, remaining)
write_matrices(file2, matrices2 + to_move)

print("Первый файл:")
for matrix in remaining:
    for row in matrix:
        print(*row)
    print()
print("Второй файл:")
for matrix in matrices2 + to_move:
    for row in matrix:
        print(*row)
    print()
```

```python
# №3
def read_matrix_pair(filename):
    with open(filename, 'r') as f:
        data = f.read().strip().split('\n\n')
    pairs = []
    for pair_str in data:
        matrices = pair_str.strip().split('\n---\n')
        if len(matrices) != 2:
            continue
        matrix1 = [list(map(int, row.split())) for row in matrices[0].split('\n')]
        matrix2 = [list(map(int, row.split())) for row in matrices[1].split('\n')]
        pairs.append((matrix1, matrix2))
    return pairs

def matrix_multiply(A, B):
    m = len(A)
    n = len(A[0])
    l = len(B[0])
    result = [[0]*l for _ in range(m)]
    for i in range(m):
        for j in range(l):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result

def write_matrices(filename, matrices):
    with open(filename, 'w') as f:
        for matrix in matrices:
            for row in matrix:
                f.write(' '.join(map(str, row)) + '\n')
            f.write('\n')

file1 = input("Введите имя первого файла: ")
file2 = input("Введите имя второго файла: ")

pairs = read_matrix_pair(file1)
products = []
for A, B in pairs:
    product = matrix_multiply(A, B)
    products.append(product)

write_matrices(file2, products)

print("Первый файл:")
for A, B in pairs:
    print("Матрица A:")
    for row in A:
        print(*row)
    print("Матрица B:")
    for row in B:
        print(*row)
    print()
print("Второй файл:")
for matrix in products:
    for row in matrix:
        print(*row)
    print()
```

```python
# №8
def read_square_matrices(filename):
    with open(filename, 'r') as f:
        data = f.read().strip().split('\n\n')
    matrices = []
    for matrix_str in data:
        rows = matrix_str.strip().split('\n')
        matrix = [list(map(int, row.split())) for row in rows]
        matrices.append(matrix)
    return matrices

def write_square_matrices(filename, matrices):
    with open(filename, 'w') as f:
        for matrix in matrices:
            for row in matrix:
                f.write(' '.join(map(str, row)) + '\n')
            f.write('\n')

def transpose(matrix):
    n = len(matrix)
    return [[matrix[j][i] for j in range(n)] for i in range(n)]

file1 = input("Введите имя первого файла: ")
file2 = input("Введите имя второго файла: ")

matrices = read_square_matrices(file1)
odd_sum_matrices = []
new_matrices = []

for matrix in matrices:
    total = 0
    n = len(matrix)
    for i in range(n):
        total += matrix[i][i]
        if i != n - i - 1:
            total += matrix[i][n - i - 1]
    if total % 2 == 1:
        odd_sum_matrices.append(matrix)
        new_matrices.append(transpose(matrix))
    else:
        new_matrices.append(matrix)

write_square_matrices(file1, new_matrices)
write_square_matrices(file2, odd_sum_matrices)

print("Первый файл:")
for matrix in new_matrices:
    for row in matrix:
        print(*row)
    print()
print("Второй файл:")
for matrix in odd_sum_matrices:
    for row in matrix:
        print(*row)
    print()
```

```python
# №9
def read_square_matrices(filename):
    with open(filename, 'r') as f:
        data = f.read().strip().split('\n\n')
    matrices = []
    for matrix_str in data:
        rows = matrix_str.strip().split('\n')
        matrix = [list(map(int, row.split())) for row in rows]
        matrices.append(matrix)
    return matrices

def write_square_matrices(filename, matrices):
    with open(filename, 'w') as f:
        for matrix in matrices:
            for row in matrix:
                f.write(' '.join(map(str, row)) + '\n')
            f.write('\n')

def is_symmetric(matrix):
    n = len(matrix)
    for i in range(n):
        for j in range(i + 1, n):
            if matrix[i][j] != matrix[j][i]:
                return False
    return True

file1 = input("Введите имя первого файла: ")
file2 = input("Введите имя второго файла: ")
file3 = input("Введите имя третьего файла: ")

matrices = read_square_matrices(file1)
symmetric = []
non_symmetric = []

for matrix in matrices:
    if is_symmetric(matrix):
        symmetric.append(matrix)
    else:
        non_symmetric.append(matrix)

write_square_matrices(file2, symmetric)
write_square_matrices(file3, non_symmetric)

print("Первый файл:")
for matrix in matrices:
    for row in matrix:
        print(*row)
    print()
print("Второй файл (симметричные):")
for matrix in symmetric:
    for row in matrix:
        print(*row)
    print()
print("Третий файл (несимметричные):")
for matrix in non_symmetric:
    for row in matrix:
        print(*row)
    print()
```

```python
# №10
def read_matrices(filename):
    with open(filename, 'r') as f:
        data = f.read().strip().split('\n\n')
    matrices = []
    for matrix_str in data:
        rows = matrix_str.strip().split('\n')
        matrix = [list(map(int, row.split())) for row in rows]
        matrices.append(matrix)
    return matrices

def write_matrix_triples(filename, triples):
    with open(filename, 'w') as f:
        for A, B, C in triples:
            f.write("Матрица A:\n")
            for row in A:
                f.write(' '.join(map(str, row)) + '\n')
            f.write("Матрица B:\n")
            for row in B:
                f.write(' '.join(map(str, row)) + '\n')
            f.write("Матрица C (произведение):\n")
            for row in C:
                f.write(' '.join(map(str, row)) + '\n')
            f.write('---\n')

def matrix_multiply(A, B):
    m = len(A)
    n = len(A[0])
    l = len(B[0])
    result = [[0]*l for _ in range(m)]
    for i in range(m):
        for j in range(l):
            for k in range(n):
                result[i][j] += A[i][k] * B[k][j]
    return result

file1 = input("Введите имя первого файла: ")
file2 = input("Введите имя второго файла: ")
file3 = input("Введите имя третьего файла: ")

matrices1 = read_matrices(file1)
matrices2 = read_matrices(file2)

triples = []
for A, B in zip(matrices1, matrices2):
    C = matrix_multiply(A, B)
    triples.append((A, B, C))

write_matrix_triples(file3, triples)

print("Первый файл:")
for matrix in matrices1:
    for row in matrix:
        print(*row)
    print()
print("Второй файл:")
for matrix in matrices2:
    for row in matrix:
        print(*row)
    print()
print("Третий файл:")
for A, B, C in triples:
    print("Матрица A:")
    for row in A:
        print(*row)
    print("Матрица B:")
    for row in B:
        print(*row)
    print("Матрица C (произведение):")
    for row in C:
        print(*row)
    print("---")
```

```python
# №12
def read_matrix_pair(filename):
    with open(filename, 'r') as f:
        data = f.read().strip().split('\n\n')
    pairs = []
    for pair_str in data:
        matrices = pair_str.strip().split('\n---\n')
        if len(matrices) != 2:
            continue
        matrix1 = [list(map(int, row.split())) for row in matrices[0].split('\n')]
        matrix2 = [list(map(int, row.split())) for row in matrices[1].split('\n')]
        pairs.append((matrix1, matrix2))
    return pairs

def matrix_add(A, B):
    m = len(A)
    n = len(A[0])
    result = [[0]*n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            result[i][j] = A[i][j] + B[i][j]
    return result

def write_matrices(filename, matrices):
    with open(filename, 'w') as f:
        for matrix in matrices:
            for row in matrix:
                f.write(' '.join(map(str, row)) + '\n')
            f.write('\n')

file1 = input("Введите имя первого файла: ")
file2 = input("Введите имя второго файла: ")

pairs = read_matrix_pair(file1)
sums = []
for A, B in pairs:
    sum_matrix = matrix_add(A, B)
    sums.append(sum_matrix)

write_matrices(file2, sums)

print("Первый файл:")
for A, B in pairs:
    print("Матрица A:")
    for row in A:
        print(*row)
    print("Матрица B:")
    for row in B:
        print(*row)
    print()
print("Второй файл:")
for matrix in sums:
    for row in matrix:
        print(*row)
    print()
```
