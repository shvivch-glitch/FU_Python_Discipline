# ЗАДАНИЕ 11. Словари.

```python
# №1
# a)
my_dict = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}
sorted_asc = dict(sorted(my_dict.items(), key=lambda x: x[1]))
sorted_desc = dict(sorted(my_dict.items(), key=lambda x: x[1], reverse=True))
print("По возрастанию:", sorted_asc)
print("По убыванию:", sorted_desc)

# б)
dict1 = {1: 2, 3: 4}
dict2 = {3: 5, 6: 7}
merged = dict1.copy()
for key, value in dict2.items():
    merged[key] = merged.get(key, 0) + value
print("Объединенный словарь:", merged)

# в)
check_keys = [1, 3]
all_exist = all(key in my_dict for key in check_keys)
print("Все ключи присутствуют:", all_exist)
```

```python
# №2
# a)
my_dict = {1: 2, 3: 4}
key = int(input("Введите ключ: "))
value = int(input("Введите значение: "))
my_dict[key] = value
print("Обновленный словарь:", my_dict)

# б)
my_dict = {1: 2, 3: 4, 5: 2}
unique_values = set(my_dict.values())
print("Уникальные значения:", unique_values)

# в)
my_dict = {1: [2, 3, 4], 3: [5, 6, 7], 5: [8, 9, 0]}
total_elements = sum(len(lst) for lst in my_dict.values())
print("Общее количество элементов:", total_elements)
```

```python
# №3
# a)
dicts = [{1: 2}, {3: 4}, {5: 6}]
merged = {}
for d in dicts:
    merged.update(d)
print("Объединенный словарь:", merged)

# б)
my_dict = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}
sorted_dict = dict(sorted(my_dict.items(), key=lambda x: x[1]))
print("Отсортированный словарь:", sorted_dict)

# в)
my_dict = {1: [2, 3, 4], 3: [5, 6, 7], 5: [8, 9, 0]}
cleaned = {key: [] for key in my_dict}
print("Очищенный словарь:", cleaned)
```

```python
# №4
# a)
my_dict = {1: 2, 3: 4}
keys_to_check = [1, 3]
all_present = all(key in my_dict for key in keys_to_check)
print("Все ключи присутствуют:", all_present)

# б)
from itertools import permutations
my_dict = {1: 'abc', 2: 'def'}
key = int(input("Введите ключ для генерации комбинаций: "))
if key in my_dict:
    combos = [''.join(p) for p in permutations(my_dict[key])]
    print("Комбинации:", combos)

# в)
my_dict = {1: [2, 3, 4], 3: [5, 6, 7], 5: [8, 9, 0]}
averages = {key: sum(values) / len(values) for key, values in my_dict.items()}
print("Средние значения:", averages)
```

```python
# №5
# a)
my_dict = {1: 2, 3: 4, 5: 6}
for key, value in my_dict.items():
    print(f"{key}: {value}")

# б)
values = sorted(my_dict.values(), reverse=True)[:3]
print("Три наибольших значения:", values)

# в)
dict1 = {1: 2, 3: 4}
dict2 = {1: 3, 3: 4}
comparison = {}
for key in set(dict1) | set(dict2):
    comparison[key] = (dict1.get(key), dict2.get(key))
print("Сравнение:", comparison)
```

```python
# №6
# a)
n = int(input("Введите n: "))
squares = {i: i**2 for i in range(1, n+1)}
print("Словарь квадратов:", squares)

# б)
dict_list = [{1: 2}, {3: 4}, {5: 6}]
values = [list(d.values())[0] for d in dict_list]
print("Объединенные значения:", values)

# в)
my_dict = {1: '', 2: 'abc', 3: None}
cleaned = {k: v for k, v in my_dict.items() if v}
print("Очищенный словарь:", cleaned)
```

```python
# №7
# a)
squares = {i: i**2 for i in range(1, 16)}
print("Словарь квадратов:", squares)

# б)
text = input("Введите строку: ")
freq = {}
for char in text:
    freq[char] = freq.get(char, 0) + 1
print("Частоты символов:", freq)

# в)
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
filtered = {k: v for k, v in my_dict.items() if v > 2}
print("Отфильтрованный словарь:", filtered)
```

```python
# №8
# a)
dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
merged = {**dict1, **dict2}
print("Объединенный словарь:", merged)

# б)
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
print("key | value")
print("----|------")
for key, value in my_dict.items():
    print(f"{key}   | {value}")

# в)
lists = [[1, 2, 3], [4, 5, 6]]
result = {f'key{i+1}': lst for i, lst in enumerate(lists)}
print("Словарь из списков:", result)
```

```python
# №9
# a)
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
for key, value in my_dict.items():
    print(f"key: {key}, value: {value}")

# б)
dict_list = [{'a': 1}, {'a': 2}, {'a': 3}, {'b': 1}]
key = 'a'
total = sum(d.get(key, 0) for d in dict_list)
print(f"Сумма значений для ключа '{key}':", total)

# в)
students = {'Иван': {'рост': 170, 'вес': 70}, 'Михаил': {'рост': 180, 'вес': 75}}
filtered = {name: {'рост': data['рост']} for name, data in students.items()}
print("Отфильтрованные данные:", filtered)
```

```python
# №10
# a)
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
total = sum(my_dict.values())
print("Сумма значений:", total)

# б)
my_list = ['a', 'b', 'c']
result = {item: {} for item in my_list}
print("Словарь из списка:", result)

# в)
my_dict = {'a': 1, 'b': 1, 'c': 1}
all_same = len(set(my_dict.values())) == 1
print("Все значения одинаковы:", all_same)
```

```python
# №11
# a)
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4}
product = 1
for value in my_dict.values():
    product *= value
print("Произведение значений:", product)

# б)
my_dict = {'a': ['c', 'b', 'a'], 'b': ['c', 'b', 'a']}
sorted_dict = {key: sorted(lst) for key, lst in my_dict.items()}
print("Отсортированные списки:", sorted_dict)

# в)
pairs = [('a', 1), ('b', 2), ('a', 3), ('b', 4)]
grouped = {}
for key, value in pairs:
    if key not in grouped:
        grouped[key] = []
    grouped[key].append(value)
print("Сгруппированные значения:", grouped)
```

```python
# №12
# a)
my_dict = {'a': 1, 'b': 2, 'c': 3}
key_to_remove = input("Введите ключ для удаления: ")
if key_to_remove in my_dict:
    del my_dict[key_to_remove]
print("Словарь после удаления:", my_dict)

# б)
my_dict = {'a ': 1, ' b': 2, 'c ': 3}
cleaned = {key.strip(): value for key, value in my_dict.items()}
print("Словарь без пробелов в ключах:", cleaned)

# в)
my_dict = {'a': [1, 2], 'b': [3, 4]}
result = []
for key, values in my_dict.items():
    for value in values:
        result.append({key: value})
print("Список словарей:", result)
```

```python
# №13
# a)
keys = ['a', 'b', 'c']
values = [1, 2, 3]
result = dict(zip(keys, values))
print("Словарь из списков:", result)

# б)
items = [
    {'name': 'item1', 'price': 10},
    {'name': 'item2', 'price': 50},
    {'name': 'item3', 'price': 30},
    {'name': 'item4', 'price': 40}
]
top_three = sorted(items, key=lambda x: x['price'], reverse=True)[:3]
print("Три самых дорогих товара:", top_three)

# в)
dict_list = [{'a': 1}, {'b': 2}, {'c': 3}]
to_remove = {'b': 2}
result = [d for d in dict_list if d != to_remove]
print("Список после удаления:", result)
```

```python
# №14
# a)
my_dict = {'b': 1, 'a': 2, 'c': 3}
sorted_dict = dict(sorted(my_dict.items()))
print("Отсортированный словарь:", sorted_dict)

# б)
for key, value in my_dict.items():
    print(f"key: {key}, value: {value}, item: ({key}, {value})")

# в)
my_dict = {'a': '1', 'b': '2', 'c': '3.0'}
converted = {}
for key, value in my_dict.items():
    try:
        converted[key] = int(value)
    except ValueError:
        converted[key] = float(value)
print("Преобразованный словарь:", converted)
```

```python
# №15
# a)
my_dict = {'a': 1, 'b': 2, 'c': 3}
max_val = max(my_dict.values())
min_val = min(my_dict.values())
print(f"Максимальное значение: {max_val}, Минимальное значение: {min_val}")

# б)
for key, value in my_dict.items():
    print(f"key: {key}, value: {value}")

# в)
my_dict = {
    'x': list(range(11, 21)),
    'y': list(range(21, 31)),
    'z': list(range(31, 41))
}
for key, values in my_dict.items():
    print(f"Пятый элемент из {key}: {values[4]}")
```
