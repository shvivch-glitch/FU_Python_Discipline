```python
# №1
# a
s = input("Введите последовательность символов: ")
digits = set(char for char in s if char.isdigit())
operators = set(char for char in s if char in '+-*/')
letters_AF = set(char for char in s if 'A' <= char <= 'F')
letters_XYZ = set(char for char in s if 'X' <= char <= 'Z')

print("Цифры:", digits)
print("Знаки операций:", operators)
print("Буквы A-F:", letters_AF)
print("Буквы X-Z:", letters_XYZ)

# b
all_products = set(range(1, 21))
print("Все коды товаров:", all_products)

input_codes = input("Введите коды товаров через пробел: ")
available_codes = set(map(int, input_codes.split()))
missing_products = all_products - available_codes

print("Отсутствующие товары:", missing_products)
```

```python
# №2
# a
user_input = input("Введите строку с цифрами и операторами: ")
digits = set(char for char in user_input if char.isdigit())
operators = set(char for char in user_input if char in '+-*/')

print("Количество уникальных цифр:", len(digits))
print("Количество уникальных операторов:", len(operators))
print("Цифры:", digits)
print("Операторы:", operators)

# b
list1 = list(map(int, input("Введите первый список чисел через пробел: ").split()))
list2 = list(map(int, input("Введите второй список чисел через пробел: ").split()))

common_elements = set(list1) & set(list2)
print("Общие элементы:", common_elements)
```

```python
# №3
# a
s = input("Введите строку: ")
letters = set(char for char in s if char.islower() and char.isalpha())
punctuation = set(char for char in s if char in '.,!?;:')
print("Уникальные строчные латинские буквы:", letters)
print("Количество знаков препинания:", len(punctuation))

# b
A = set(input("Введите элементы множества A через пробел: ").split())
B = set(input("Введите элементы множества B через пробел: ").split())
print("A является подмножеством B:", A.issubset(B))
```

```python
# №4
# a
s1 = input("Введите первое предложение: ")
s2 = input("Введите второе предложение: ")
s3 = input("Введите третье предложение: ")
common = set(s1) & set(s2) & set(s3)
print("Общие буквы для трех предложений:", common)

# b
uni1 = set(input("Введите предметы первого университета через пробел: ").split())
uni2 = set(input("Введите предметы второго университета через пробел: ").split())
unique1 = uni1 - uni2
unique2 = uni2 - uni1
print("Предметы, изучаемые только в одном университете:", unique1 | unique2)
```

```python
# №5
# a
a = int(input("Введите первое число: "))
b = int(input("Введите второе число: "))
c = int(input("Введите третье число: "))
digits = set(str(a) + str(b) + str(c))
max_digit = max(digits)
print("Наибольшая цифра:", max_digit)

# b
A = set(input("Введите элементы множества A через пробел: ").split())
B = set(input("Введите элементы множества B через пробел: ").split())
symmetric_diff = A ^ B
print("Элементы, присутствующие только в одном из множеств:", symmetric_diff)
```

```python
# №6
# a
s = input("Введите строку: ")
unique_digits = set(char for char in s if char.isdigit())
print("Количество уникальных цифр:", len(unique_digits))
print("Уникальные цифры:", unique_digits)

# b
user1 = set(input("Введите товары первого пользователя через пробел: ").split())
user2 = set(input("Введите товары второго пользователя через пробел: ").split())
unique_to_one = user1 ^ user2
print("Товары, купленные только одним пользователем:", unique_to_one)
```

```python
# №7
# a
s1 = input("Введите первое предложение: ")
s2 = input("Введите второе предложение: ")
s3 = input("Введите третье предложение: ")
unique1 = set(s1) - set(s2) - set(s3)
unique2 = set(s2) - set(s1) - set(s3)
unique3 = set(s3) - set(s1) - set(s2)
print("Уникальные буквы для трех предложений:", unique1 | unique2 | unique3)

# b
A = set(input("Введите элементы множества A через пробел: ").split())
B = set(input("Введите элементы множества B через пробел: ").split())
print("Множества не пересекаются:", A.isdisjoint(B))
```

```python
# №8
# a
s1 = input("Введите первую строку: ")
s2 = input("Введите вторую строку: ")
target = input("Введите целевую строку: ")
available_chars = set(s1 + s2)
print("Можно составить третью строку:", set(target).issubset(available_chars))

# b
n = int(input("Введите количество сотрудников: "))
skills_dict = {}
for i in range(n):
    name = input(f"Введите имя сотрудника {i+1}: ")
    skills = set(input(f"Введите навыки сотрудника {i+1} через пробел: ").split())
    skills_dict[name] = skills

all_skills = set().union(*skills_dict.values())
common_skills = set().intersection(*skills_dict.values())
unique_skills = all_skills - common_skills
print("Навыки, которые есть только у одного сотрудника:", unique_skills)
```

```python
# №9
# a
text = input("Введите текст: ").lower()
vowels = set('аеёиоуыэюя')
found = set(char for char in text if char in vowels)
print("Уникальные русские гласные буквы:", sorted(found))

# b
n = int(input("Введите количество продуктов: "))
categories = {}
for i in range(n):
    product = input(f"Введите название продукта {i+1}: ")
    cat_list = input(f"Введите категории продукта {i+1} через пробел: ").split()
    for cat in cat_list:
        categories.setdefault(cat, set()).add(product)

unique_cats = [cat for cat, prods in categories.items() if len(prods) == 1]
print("Категории с только одним продуктом:", unique_cats)
```

```python
# №10
# a
num = input("Введите число: ")
digits = set(num)
print("Уникальные цифры в числе:", sorted(digits))

# b
A = set(input("Введите элементы множества A через пробел: ").split())
B = set(input("Введите элементы множества B через пробел: ").split())
print("A является строгим подмножеством B:", A.issubset(B) and A != B)
```

```python
# №11
# a
text = input("Введите текст: ").lower()
vowels = set('аеёиоуыэюя')
consonants = set('бвгджзйклмнпрстфхцчшщ')
vowel_count = sum(1 for char in text if char in vowels)
consonant_count = sum(1 for char in text if char in consonants)
print("Гласных букв:", vowel_count)
print("Согласных букв:", consonant_count)
print("Больше гласных" if vowel_count > consonant_count else "Больше согласных")

# b
n = int(input("Введите количество дней: "))
orders = {}
for i in range(n):
    day = input(f"Введите день {i+1}: ")
    products = set(input(f"Введите товары для дня {i+1} через пробел: ").split())
    orders[day] = products

target_day = input("Введите день для поиска уникальных товаров: ")
other_days = set().union(*[products for day, products in orders.items() if day != target_day])
unique_to_day = orders[target_day] - other_days
print(f"Товары, заказанные только в день {target_day}:", unique_to_day)
```

```python
# №12
# a
num = input("Введите натуральное число: ")
all_digits = set('0123456789')
used_digits = set(num)
missing = all_digits - used_digits
print("Цифры, не встречающиеся в числе:", sorted(missing))

# b
A = set(input("Введите элементы множества A через пробел: ").split())
B = set(input("Введите элементы множества B через пробел: ").split())
print("A является надмножеством B:", B.issubset(A))
```

```python
# №13
# a
A = set(input("Введите элементы множества A через пробел: ").split())
B = set(input("Введите элементы множества B через пробел: ").split())
A -= B
print("Множество A после удаления общих элементов:", A)

# b
n = int(input("Введите количество пользователей: "))
all_products = set()
for i in range(n):
    products = set(input(f"Введите товары пользователя {i+1} через пробел: ").split())
    all_products |= products
print("Товары, купленные хотя бы одним пользователем:", all_products)
```

```python
# №14
# a
text = input("Введите текст: ").lower()
all_consonants = set('бвгджзйклмнпрстфхцчшщ')
used_letters = set(text)
missing = all_consonants - used_letters
print("Согласные буквы, не встречающиеся в тексте:", sorted(missing))

# b
A = set(input("Введите элементы множества A через пробел: ").split())
B = set(input("Введите элементы множества B через пробел: ").split())
print("Множества равны:", A == B)
```
