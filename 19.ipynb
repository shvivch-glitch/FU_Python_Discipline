# ЗАДАНИЕ 19. Файлы.

```python
# №1

k = int(input("Введите K: "))
input_filename = input("Введите имя исходного файла: ")
output_filename = input("Введите имя выходного файла: ")

with open(input_filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

if len(lines) > k:
    last_k_lines = lines[-k:]
    with open(output_filename, 'w', encoding='utf-8') as f:
        f.writelines(last_k_lines)
    print(f"Создан файл {output_filename} с последними {k} строками.")
else:
    print(f"Файл содержит {len(lines)} строк, что меньше или равно K={k}.")
```

```python
# №2

k = int(input("Введите K: "))
filename = input("Введите имя файла: ")

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

if len(lines) > k:
    lines = lines[:-k]
    with open(filename, 'w', encoding='utf-8') as f:
        f.writelines(lines)
    print(f"Удалено последних {k} строк из файла.")
else:
    print(f"Файл содержит {len(lines)} строк, что меньше или равно K={k}.")
```

```python
# №3

k = int(input("Введите K: "))
filename = input("Введите имя файла: ")

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

new_lines = []
for line in lines:
    if len(line) > k:
        new_lines.append(line[k:])
    else:
        new_lines.append('')

with open(filename, 'w', encoding='utf-8') as f:
    f.writelines(new_lines)

print(f"Из каждой строки удалено первые {k} символов.")
```

```python
# №4

file1 = input("Введите имя первого файла: ")
file2 = input("Введите имя второго файла: ")

with open(file1, 'r', encoding='utf-8') as f1:
    lines1 = f1.readlines()

with open(file2, 'r', encoding='utf-8') as f2:
    lines2 = f2.readlines()

result_lines = []
for i in range(len(lines1)):
    if i < len(lines2):
        line1 = lines1[i].rstrip('\n')
        line2 = lines2[i].rstrip('\n')
        result_lines.append(line1 + line2 + '\n')
    else:
        result_lines.append(lines1[i])

with open(file1, 'w', encoding='utf-8') as f1:
    f1.writelines(result_lines)

print("Строки второго файла добавлены в конец строк первого файла.")
```

```python
# №5

k = int(input("Введите K (номер строки для удаления): "))
filename = input("Введите имя файла: ")

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

if 1 <= k <= len(lines):
    del lines[k-1]
    with open(filename, 'w', encoding='utf-8') as f:
        f.writelines(lines)
    print(f"Строка с номером {k} удалена.")
else:
    print(f"Строки с номером {k} нет в файле. Файл не изменен.")
```

```python
# №6

import string

input_filename = input("Введите имя текстового файла: ")
output_filename = input("Введите имя символьного файла: ")

punctuation_chars = []
with open(input_filename, 'r', encoding='utf-8') as f:
    content = f.read()
    for char in content:
        if char in string.punctuation:
            punctuation_chars.append(char)

with open(output_filename, 'w', encoding='utf-8') as f:
    f.write(''.join(punctuation_chars))

print(f"Создан файл {output_filename} со знаками препинания.")
```

```python
# №7

filename = input("Введите имя файла: ")

numbers = []
with open(filename, 'r', encoding='utf-8') as f:
    for line in f:
        stripped_line = line.strip()
        if stripped_line:
            try:
                numbers.append(int(stripped_line))
            except ValueError:
                continue

count = len(numbers)
total_sum = sum(numbers)

print(f"Количество чисел: {count}")
print(f"Сумма чисел: {total_sum}")
```

```python
# №8

s = input("Введите строку S: ")
filename = input("Введите имя файла: ")

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

for i in range(len(lines)):
    if lines[i].strip() == '':
        lines[i] = s + '\n'

with open(filename, 'w', encoding='utf-8') as f:
    f.writelines(lines)

print("Пустые строки заменены на строку S.")
```

```python
# №9

filename = input("Введите имя файла: ")
width = 50

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

centered_lines = []
for line in lines:
    stripped = line.rstrip('\n')
    if stripped:
        if len(stripped) % 2 == 1:
            stripped = ' ' + stripped
        padding = (width - len(stripped)) // 2
        centered = ' ' * padding + stripped
        centered_lines.append(centered + '\n')
    else:
        centered_lines.append('\n')

with open(filename, 'w', encoding='utf-8') as f:
    f.writelines(centered_lines)

print("Текст выровнен по центру.")
```

```python
# №10

filename = input("Введите имя файла: ")

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

paragraph_count = 0
in_paragraph = False

for line in lines:
    if line.strip() == '':
        if in_paragraph:
            in_paragraph = False
    else:
        if not in_paragraph:
            paragraph_count += 1
            in_paragraph = True

print(f"Количество абзацев: {paragraph_count}")
```

```python
# №11

filename = input("Введите имя файла: ")
width = 50

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

right_aligned_lines = []
for line in lines:
    stripped = line.rstrip('\n')
    if stripped:
        padding = width - len(stripped)
        if padding > 0:
            right_aligned = ' ' * padding + stripped
        else:
            right_aligned = stripped
        right_aligned_lines.append(right_aligned + '\n')
    else:
        right_aligned_lines.append('\n')

with open(filename, 'w', encoding='utf-8') as f:
    f.writelines(right_aligned_lines)

print("Текст выровнен по правому краю.")
```

```python
# №12

file1 = input("Введите имя первого файла: ")
file2 = input("Введите имя второго файла: ")

with open(file2, 'r', encoding='utf-8') as f2:
    content2 = f2.read()

with open(file1, 'r', encoding='utf-8') as f1:
    content1 = f1.read()

with open(file1, 'w', encoding='utf-8') as f1:
    f1.write(content2 + content1)

print(f"Содержимое файла {file2} добавлено в начало файла {file1}.")
```

```python
# №13

k = int(input("Введите K (номер строки перед которой вставить пустую строку): "))
filename = input("Введите имя файла: ")

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

if 1 <= k <= len(lines):
    lines.insert(k-1, '\n')
    with open(filename, 'w', encoding='utf-8') as f:
        f.writelines(lines)
    print(f"Пустая строка вставлена перед строкой {k}.")
else:
    print(f"Строки с номером {k} нет в файле. Файл не изменен.")
```

```python
# №14

filename = input("Введите имя файла: ")

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

new_lines = []
for line in lines:
    new_lines.append(line)
    if line.strip() == '':
        new_lines.append(line)

with open(filename, 'w', encoding='utf-8') as f:
    f.writelines(new_lines)

print("Все пустые строки продублированы.")
```

```python
# №15

filename = input("Введите имя файла: ")

with open(filename, 'r', encoding='utf-8') as f:
    lines = f.readlines()

if len(lines) > 2:
    lines = lines[1:-1]
    with open(filename, 'w', encoding='utf-8') as f:
        f.writelines(lines)
    print("Первая и последняя строки удалены.")
else:
    print("Файл содержит 2 или меньше строк. Нечего удалять.")
```
