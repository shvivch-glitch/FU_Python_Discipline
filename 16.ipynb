# ЗАДАНИЕ 16. Функции: необязательные параметры, неопределенное количество параметров


```python
# №1
# a
def full_name(first_name, last_name, middle_name=None):
    if middle_name:
        return f"{last_name} {first_name} {middle_name}"
    return f"{last_name} {first_name}"

# b
def transform_list(numbers, transform_function=None):
    if transform_function:
        return [transform_function(x) for x in numbers]
    return numbers.copy()

# c
def sum_positive_numbers(*args):
    return sum(x for x in args if x > 0)

# d
def common_keys(**kwargs):
    dicts = list(kwargs.values())
    common = dicts[0]
    for d in dicts[1:]:
        common = {k: d[k] for k in common if k in d}
    return common

print()
```

```python
# №2
# a
def sort_numbers(numbers, reverse_order=False):
    return sorted(numbers, reverse=reverse_order)

# b
def uppercase_text(text, separator=' '):
    return separator.join(word.upper() for word in text.split(separator))

# c
def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

# d
def filter_strings(strings, **kwargs):
    result = []
    for s in strings:
        valid = True
        if 'min_length' in kwargs and len(s) < kwargs['min_length']:
            valid = False
        if 'starts_with' in kwargs and not s.startswith(kwargs['starts_with']):
            valid = False
        if 'ends_with' in kwargs and not s.endswith(kwargs['ends_with']):
            valid = False
        if valid:
            result.append(s)
    return result

print()
```

```python
# №3
# a
def split_string(text, delimiter=' '):
    return text.split(delimiter)

# b
def compare_elements(list1, list2, comparison_function=None):
    if comparison_function is None:
        comparison_function = lambda x, y: x == y
    return [(x, y) for x, y in zip(list1, list2) if comparison_function(x, y)]

# c
def count_unique_chars(*args):
    return len(set(''.join(args)))

# d
def common_elements(*lists):
    return list(set.intersection(*map(set, lists)))

print()
```

```python
# №4
# a
def calculate(a, b, operation='add'):
    operations = {
        'add': a + b,
        'subtract': a - b,
        'multiply': a * b,
        'divide': a / b
    }
    return operations[operation]

# b
def modify_strings(strings, case=None):
    if case == 'upper':
        return [s.upper() for s in strings]
    elif case == 'lower':
        return [s.lower() for s in strings]
    return strings

# c
def average(*args):
    return sum(args) / len(args)

# d
def merge_dictionaries(base_dict, **kwargs):
    result = base_dict.copy()
    for d in kwargs.values():
        result.update(d)
    return result

print()
```

```python
# №5
# a
def filter_items(items, condition=None):
    if condition:
        return [item for item in items if condition(item)]
    return items

# b
def merge_and_filter(list1, list2, filter_function=None):
    merged = list1 + list2
    if filter_function:
        return [item for item in merged if filter_function(item)]
    return merged

# c
def common_in_all(*lists):
    return list(set.intersection(*map(set, lists)))

# d
def find_primes(*numbers):
    def is_prime(n):
        if n < 2:
            return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True
    return [n for n in numbers if is_prime(n)]

print()
```

```python
# №6
# a
def word_count(text, separator=' '):
    words = text.split(separator)
    return {word: words.count(word) for word in set(words)}

# b
def filter_numbers(numbers, exclude_multiples_of=2):
    return [n for n in numbers if n % exclude_multiples_of != 0]

# c
def merge_and_sort_keys(*dicts):
    merged = {}
    for d in dicts:
        for k, v in d.items():
            if k not in merged or v > merged[k]:
                merged[k] = v
    return sorted(merged.keys(), key=lambda k: merged[k], reverse=True)

# d
def sum_non_negative(*args):
    return sum(x for x in args if x >= 0)

print()
```

```python
# №7
# a
def join_strings(strings, delimiter=','):
    return delimiter.join(strings)

# b
def combine_lists(list1, list2, join_str=''):
    return [f"{x}{join_str}{y}" for x, y in zip(list1, list2)]

# c
def unique_values(*args):
    return list(set(args))

# d
def reverse_concatenate(*args):
    return ' '.join(reversed(args))

print()
```

```python
# №8
# a
def pair_and_filter(list1, list2, filter_function=None):
    pairs = [(x, y) for x in list1 for y in list2]
    if filter_function:
        return [p for p in pairs if filter_function(*p)]
    return pairs

# b
def concat_or_upper(strings, uppercase=False):
    result = ' '.join(strings)
    return result.upper() if uppercase else result

# c
def filter_uppercase_strings(strings):
    return [s for s in strings if s.isupper()]

# d
def unique_sorted_elements(*lists):
    return sorted(set().union(*lists))

print()
```

```python
# №9
# a
def capitalize_words(text, separator=' '):
    words = text.split(separator)
    return separator.join(word.capitalize() for word in words)

# b
def filter_elements(list1, list2, filter_function=None):
    combined = list1 + list2
    if filter_function:
        return [item for item in combined if filter_function(item)]
    return combined

# c
def merge_dictionaries(*dicts):
    result = {}
    for d in dicts:
        result.update(d)
    return result

# d
def unique_keys(*dicts):
    all_keys = {}
    for d in dicts:
        for k, v in d.items():
            if k not in all_keys:
                all_keys[k] = v
            else:
                all_keys[k] = None
    return {k: v for k, v in all_keys.items() if v is not None}

print()
```

```python
# №10
# a
def filter_numbers(numbers, condition=None):
    if condition:
        return [n for n in numbers if condition(n)]
    return numbers

# b
def transform_numbers(numbers, transform_function=None):
    if transform_function:
        return [transform_function(n) for n in numbers]
    return numbers

# c
def cube_numbers(*lists):
    return [x**3 for sublist in lists for x in sublist]

# d
def product_non_zero(*args):
    result = 1
    for num in args:
        if num != 0:
            result *= num
    return result

print()
```

```python
# №11
# a
def format_strings(strings, separator=' '):
    return separator.join(s.capitalize() for s in strings)

# b
def merge_and_sort(list1, list2, sort_function=None):
    merged = list1 + list2
    return sorted(merged, key=sort_function) if sort_function else sorted(merged)

# c
def concat_all_strings(*lists):
    all_strings = [s for sublist in lists for s in sublist]
    return ','.join(all_strings)

# d
def unique_sorted_strings(*args):
    return sorted(set(args))

print()
```

```python
# №12
# a
def combine_lists(list1, list2, order='first_second'):
    return list1 + list2 if order == 'first_second' else list2 + list1

# b
def process_string(text, separator=None):
    if separator:
        words = text.split(separator)
        return separator.join(word.capitalize() for word in words)
    return text.lower()

# c
def difference(first, *others):
    others_combined = set().union(*others)
    return [item for item in first if item not in others_combined]

# d
def common_in_at_least_two(*lists):
    from collections import Counter
    all_elements = [element for sublist in lists for element in set(sublist)]
    count = Counter(all_elements)
    return [element for element, c in count.items() if c >= 2]

print()
```

```python
# №13
# a
def divisible_by(numbers, divisor=2):
    return [n for n in numbers if n % divisor == 0]

# b
def transform_numbers(numbers, transform_function=None):
    if transform_function:
        return [transform_function(n) for n in numbers]
    return [n for n in numbers if n >= 0]

# c
def sorted_values_by_keys(*dicts):
    merged = {}
    for d in dicts:
        merged.update(d)
    return [v for k, v in sorted(merged.items(), key=lambda x: int(x[0]))]

# d
def common_keys_in_dicts(*dicts):
    common = set(dicts[0].keys())
    for d in dicts[1:]:
        common &= set(d.keys())
    return {k: dicts[-1][k] for k in common}

print()
```

```python
# №14
# a
def merge_lists_with(list1, list2, merge_function=None):
    if merge_function:
        return [merge_function(x, y) for x, y in zip(list1, list2)]
    return list1 + list2

# b
def extract_digits(strings, digits_only=False):
    if digits_only:
        return [''.join(c for c in s if c.isdigit()) for s in strings]
    return strings

# c
def filter_even_numbers(*args):
    return [n for n in args if n % 2 == 0]

# d
def sorted_positive_numbers(*args):
    positive = [n for n in args if n > 0]
    return sorted(positive, reverse=True)

print()
```

```python
# №15
# a
def change_case(strings, to_upper=False):
    result = ' '.join(strings)
    return result.upper() if to_upper else result.lower()

# b
def filter_combined_lists(list1, list2, filter_function=None):
    combined = list1 + list2
    if filter_function:
        return [item for item in combined if filter_function(item)]
    return combined

# c
def unique_sorted_numbers(numbers):
    return sorted(set(numbers))

# d
def strings_starting_with_upper(*args):
    return [s for s in args if s and s[0].isupper()]

print()
```
