```python
# №9
x1 = float(input("Введите x1: "))
y1 = float(input("Введите y1: "))
x2 = float(input("Введите x2: "))
y2 = float(input("Введите y2: "))
x3 = float(input("Введите x3: "))
y3 = float(input("Введите y3: "))

def distance(ax, ay, bx, by):
    return ((ax - bx) ** 2 + (ay - by) ** 2) ** 0.5

a = distance(x2, y2, x3, y3)
b = distance(x1, y1, x3, y3)
c = distance(x1, y1, x2, y2)

sides = sorted([a, b, c])
if abs(a - b) < 1e-9 and abs(b - c) < 1e-9:
    print("равносторонний")
elif abs(a - b) < 1e-9 or abs(b - c) < 1e-9 or abs(a - c) < 1e-9:
    print("равнобедренный")
elif abs(sides[0]**2 + sides[1]**2 - sides[2]**2) < 1e-9:
    print("прямоугольный")
else:
    print("обычный")
```

```python
# №12
A = float(input("Введите A: "))
B = float(input("Введите B: "))
Op = int(input("Введите операцию (1-вычитание, 2-умножение, 3-деление, иное-сложение): "))
if Op == 1:
    result = A - B
elif Op == 2:
    result = A * B
elif Op == 3:
    result = A / B
else:
    result = A + B
print(f"Результат: {result}")
```

```python
# №13
def LuckyNum(num):
    s = str(num).zfill(4)
    return int(s[0]) + int(s[1]) == int(s[2]) + int(s[3])

print("Счастливые номера:")
for n in range(10000):
    if LuckyNum(n):
        print(str(n).zfill(4))
```
