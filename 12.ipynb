# ЗАДАНИЕ 12. Кортежи.

```python
# №1
# a
n = int(input("Введите количество строк: "))
strings = []
for i in range(n):
    s = input(f"Введите строку {i+1}: ")
    strings.append(s)
t = tuple(strings)
print("Созданный кортеж:", t)

# b
result_dict = {}
for s in t:
    upper_count = sum(1 for char in s if char.isupper())
    result_dict[s] = upper_count
print("Словарь (строка: количество заглавных букв):", result_dict)
```

```python
# №2
# a
mixed_data = []
mixed_data.append(input("Введите строку: "))
mixed_data.append(int(input("Введите целое число: ")))
mixed_data.append(float(input("Введите дробное число: ")))
mixed_data.append(eval(input("Введите список в формате [1,2,3]: ")))
mixed_data.append(eval(input("Введите кортеж в формате (4,5,6): ")))

t = tuple(mixed_data)
print("Кортеж с разными типами данных:", t)

# b
n = int(input("Введите количество кортежей: "))
list_of_tuples = []
for i in range(n):
    print(f"Кортеж {i+1}:")
    elements = []
    elements.append(input("Введите строку: "))
    elements.append(int(input("Введите число: ")))
    elements.append(eval(input("Введите список: ")))
    list_of_tuples.append(tuple(elements))

strings = []
numbers = []
lists = []

for tup in list_of_tuples:
    for item in tup:
        if isinstance(item, str):
            strings.append(item)
        elif isinstance(item, (int, float)):
            numbers.append(item)
        elif isinstance(item, list):
            lists.append(item)

print("Строки:", strings)
print("Числа:", numbers)
print("Списки:", lists)
```

```python
# №3
# a
numbers = input("Введите числа через пробел: ").split()
t = tuple(map(int, numbers))
print("Первый элемент кортежа:", t[0])

# b
filtered = tuple(x for x in reversed(t) if x % 3 != 0)
print("Кортеж в обратном порядке без чисел, кратных 3:", filtered)
```

```python
# №4
# a
print("Введите 3 элемента кортежа:")
a = input("Элемент 1: ")
b = input("Элемент 2: ")
c = input("Элемент 3: ")
t = (a, b, c)
var1, var2, var3 = t
print("Распакованные переменные:", var1, var2, var3)

# b
n = int(input("Введите количество пар ключ-значение: "))
pairs = []
for i in range(n):
    key = input(f"Ключ {i+1}: ")
    value = int(input(f"Значение {i+1}: "))
    pairs.append((key, value))

d = dict(sorted(pairs, key=lambda x: x[1]))
print("Отсортированный словарь:", d)
```

```python
# №5
# a
initial = tuple(input("Введите элементы начального кортежа через пробел: ").split())
new_element = input("Введите новый элемент: ")
t = initial + (new_element,)
print("Обновленный кортеж:", t)

# b
reversed_elements = tuple(s[::-1] for s in t)
print("Кортеж с обратными строками:", reversed_elements)
```

```python
# №6
# a
elements = input("Введите элементы кортежа через пробел: ").split()
t = tuple(elements)
separator = input("Введите разделитель: ")
result = separator.join(t)
print("Результирующая строка:", result)

# b
n = int(input("Введите количество кортежей: "))
list_of_tuples = []
for i in range(n):
    elements = input(f"Введите элементы кортежа {i+1} через пробел: ").split()
    processed_elements = []
    for elem in elements:
        if elem.isdigit() or (elem[0] == '-' and elem[1:].isdigit()):
            processed_elements.append(int(elem))
        elif elem.replace('.', '').isdigit() or (elem[0] == '-' and elem[1:].replace('.', '').isdigit()):
            processed_elements.append(float(elem))
        else:
            processed_elements.append(elem)
    list_of_tuples.append(tuple(processed_elements))

new_value = input("Введите новое значение для замены: ")
result = []
for tup in list_of_tuples:
    if isinstance(tup[-1], (int, float)):
        result.append(tup[:-1] + (new_value,))
    else:
        result.append(tup)

print("Обновленный список кортежей:", result)
```

```python
# №7
# a
elements = input("Введите элементы кортежа через пробел: ").split()
t = tuple(elements)
if len(t) >= 4:
    print("Четвертый элемент с начала:", t[3])
    print("Четвертый элемент с конца:", t[-4])
else:
    print("Кортеж слишком короткий")

# b
import random
n = int(input("Введите количество кортежей: "))
list_of_tuples = []
for i in range(n):
    elements = input(f"Введите элементы кортежа {i+1} через пробел: ").split()
    if elements:
        list_of_tuples.append(tuple(elements))
    else:
        list_of_tuples.append(tuple())

result = []
for tup in list_of_tuples:
    if len(tup) == 0:
        result.append((random.randint(1, 100), random.randint(1, 100)))
    else:
        result.append(tup)

print("Обновленный список кортежей:", result)
```

```python
# №8
# a
n = int(input("Введите количество словарей: "))
tuple_of_dicts = []
for i in range(n):
    key = input(f"Введите ключ для словаря {i+1}: ")
    value = float(input(f"Введите числовое значение для словаря {i+1}: "))
    tuple_of_dicts.append({key: value})

print("Кортеж словарей:", tuple_of_dicts)

# b
sorted_tuple = sorted(tuple_of_dicts, key=lambda d: list(d.values())[0], reverse=True)
print("Отсортированный кортеж:", sorted_tuple)
```

```python
# №9
# a
elements = input("Введите элементы кортежа через пробел: ").split()
t = tuple(elements)
repeats = []
for item in set(t):
    if t.count(item) > 1:
        repeats.append(item)
print("Повторяющиеся элементы:", repeats)

# b
elements = input("Введите элементы списка через пробел: ").split()
processed_elements = []
for elem in elements:
    if elem.startswith('(') and elem.endswith(')'):
        processed_elements.append(tuple(elem[1:-1].split(',')))
    else:
        processed_elements.append(elem)

total = 0
for item in processed_elements:
    if isinstance(item, tuple):
        break
    if isinstance(item, str) and item.isdigit():
        total += int(item)
print("Сумма элементов до первого кортежа:", total)
```

```python
# №10
# a
elements = input("Введите элементы кортежа через пробел: ").split()
t = tuple(elements)
search = input("Введите элемент для поиска: ")
count = t.count(search)
print(f"Элемент '{search}' встречается {count} раз(а)")

# b
words = input("Введите строки через пробел: ").split()
t = tuple(dict.fromkeys(words))
print("Кортеж без повторений:", t)
```

```python
# №11
# a
elements = input("Введите элементы списка через пробел: ").split()
t = tuple(elements)
print("Преобразованный кортеж:", t)

# b
numbers = [int(x) for x in t if x.isdigit()]
product = 1
for num in numbers:
    if num % 2 == 0:
        product *= num
print("Произведение четных чисел:", product if product != 1 else 0)
```

```python
# №12
# a
elements = input("Введите элементы кортежа через пробел: ").split()
t = tuple(elements)
to_remove = input("Введите элемент для удаления: ")
new_t = tuple(x for x in t if x != to_remove)
print("Кортеж после удаления:", new_t)

# b
n = int(input("Введите количество кортежей: "))
nested = []
for i in range(n):
    elements = input(f"Введите числа для кортежа {i+1} через пробел: ").split()
    nested.append(tuple(map(int, elements)))

averages = []
for tup in nested:
    numbers = [x for x in tup if x > 5]
    if numbers:
        avg = sum(numbers) / len(numbers)
        averages.append(avg)
    else:
        averages.append(0)

print("Средние значения чисел больше 5:", averages)
```

```python
# №13
# a
elements = input("Введите элементы кортежа через пробел: ").split()
t = tuple(elements)
start = int(input("Введите начальный индекс: "))
end = int(input("Введите конечный индекс: "))
sliced = t[start:end]
filtered = tuple(x for x in sliced if not x.isalpha())
print("Срез без строк:", filtered)

# b
elements = input("Введите числа через пробел: ").split()
t = tuple(int(x) for x in elements if int(x) >= 0)
print("Кортеж без отрицательных чисел:", t)
```

```python
# №14
# a
elements = input("Введите элементы кортежа через пробел: ").split()
t = tuple(elements)
search = input("Введите число для поиска: ")
indices = [i for i, x in enumerate(t) if x == search and x.isdigit()]
print("Индексы найденного числа:", indices)

# b
numbers = input("Введите числа через пробел: ").split()
filtered = [x for x in numbers if int(x) % 3 == 0]
result = int(''.join(filtered)) if filtered else 0
print("Составленное число:", result)
```
