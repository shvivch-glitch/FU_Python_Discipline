```python
# №11

# a
from collections import Counter
text = "apple banana apple orange banana apple"
word_counts = Counter(text.split())
most_common_word = word_counts.most_common(1)[0]
print(most_common_word)

# б
from collections import defaultdict
def odd_after_even(numbers):
    dd = defaultdict(list)
    for i in range(len(numbers) - 1):
        if numbers[i] % 2 == 0 and numbers[i + 1] % 2 == 1:
            dd[numbers[i]].append(numbers[i + 1])
    return dict(dd)
numbers = [2, 3, 4, 5, 6, 7, 8, 9]
print(odd_after_even(numbers))

# в
from enum import Enum
class ColorRGB(Enum):
    RED = 1
    GREEN = 2
    BLUE = 3
def rgb_code(color):
    codes = {
        ColorRGB.RED: (255, 0, 0),
        ColorRGB.GREEN: (0, 255, 0),
        ColorRGB.BLUE: (0, 0, 255)
    }
    return codes[color]
print(rgb_code(ColorRGB.RED))

# г
def intersection_frozenset(set1, set2):
    return frozenset(set1 & set2)
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print(intersection_frozenset(set1, set2))

# д
from collections import namedtuple
Student = namedtuple('Student', ['name', 'grades'])
def high_achievers(students, threshold):
    result = []
    for student in students:
        avg = sum(student.grades) / len(student.grades)
        if avg > threshold:
            result.append(student)
    return result
students = [
    Student('Alice', [5, 4, 5]),
    Student('Bob', [3, 3, 4]),
    Student('Charlie', [5, 5, 5])
]
print(high_achievers(students, 4.0))

# е
from collections import OrderedDict
class OrderedDictManager:
    def __init__(self):
        self.od = OrderedDict()

    def add(self, key, value):
        self.od[key] = value

    def remove(self, key):
        if key in self.od:
            del self.od[key]

    def update(self, key, value):
        if key in self.od:
            self.od[key] = value

    def get_keys(self):
        return list(self.od.keys())

    def get_values(self):
        return list(self.od.values())

    def get_sorted_keys(self):
        return sorted(self.od.keys())

    def get_sorted_values(self):
        return sorted(self.od.values())

manager = OrderedDictManager()
manager.add('a', 1)
manager.add('b', 3)
manager.add('c', 2)
print(manager.get_keys())
print(manager.get_sorted_keys())
print(manager.get_sorted_values())
```

```python
# №12

# a
from collections import Counter
list12 = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
counter = Counter(list12)
least_common = counter.most_common()[-1]
print(least_common)

# б
from collections import defaultdict
def population_ranges_dict(cities):
    dd = defaultdict(list)
    for city, pop in cities:
        if pop < 100000:
            key = 'менее 100 тыс.'
        elif pop < 1000000:
            key = 'от 100 тыс. до 1 млн.'
        else:
            key = 'более 1 млн.'
        dd[key].append(city)
    return dict(dd)
cities = [('Moscow', 12000000), ('SPb', 5000000), ('Novosibirsk', 1600000), ('Tver', 400000)]
print(population_ranges_dict(cities))

# в
from enum import Enum
class AnimalWeight(Enum):
    CAT = 1
    DOG = 2
    RABBIT = 3
def average_weight(animal):
    weights = {
        AnimalWeight.CAT: 4.5,
        AnimalWeight.DOG: 15.0,
        AnimalWeight.RABBIT: 1.5
    }
    return weights[animal]
print(average_weight(AnimalWeight.CAT))

# г
def unique_keys_frozenset(dicts_list):
    keys = set()
    for d in dicts_list:
        keys.update(d.keys())
    return frozenset(keys)
dicts = [{'a': 1, 'b': 2}, {'b': 3, 'c': 4}, {'a': 5, 'd': 6}]
print(unique_keys_frozenset(dicts))

# д
from collections import namedtuple
Book = namedtuple('Book', ['title', 'author', 'pages'])
def books_with_many_pages(books, min_pages):
    return [book for book in books if book.pages > min_pages]
books = [
    Book('Book1', 'Author1', 300),
    Book('Book2', 'Author2', 150),
    Book('Book3', 'Author3', 500)
]
print(books_with_many_pages(books, 200))

# е
from collections import OrderedDict
def filter_by_value(od, min_value):
    new_od = OrderedDict()
    for key, value in od.items():
        if value > min_value:
            new_od[key] = value
    return new_od
od = OrderedDict([('a', 10), ('b', 5), ('c', 15), ('d', 3)])
filtered = filter_by_value(od, 8)
print(filtered)
```

```python
# №13

# a
from collections import Counter
text = "apple banana apple orange banana apple grape"
word_counts = Counter(text.split())
least_common_word = word_counts.most_common()[-1]
print(least_common_word)

# б
from collections import defaultdict
def price_ranges_dict(products):
    dd = defaultdict(list)
    for name, price in products:
        if price < 100:
            key = 'до 100 руб.'
        elif price < 500:
            key = 'от 100 руб. до 500 руб.'
        else:
            key = 'более 500 руб.'
        dd[key].append(name)
    return dict(dd)
products = [('apple', 50), ('banana', 120), ('orange', 80), ('milk', 600), ('bread', 30)]
print(price_ranges_dict(products))

# в
from enum import Enum
class Drink(Enum):
    TEA = 1
    COFFEE = 2
    JUICE = 3
def caffeine_content(drink):
    caffeine = {
        Drink.TEA: 30,
        Drink.COFFEE: 95,
        Drink.JUICE: 0
    }
    return caffeine[drink]
print(caffeine_content(Drink.COFFEE))

# г
def difference_second_only(set1, set2):
    return frozenset(set2 - set1)
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print(difference_second_only(set1, set2))

# д
from collections import namedtuple
Supplier = namedtuple('Supplier', ['name', 'city'])
def suppliers_by_city(suppliers, target_city):
    return [s for s in suppliers if s.city == target_city]
suppliers = [
    Supplier('Supplier1', 'Moscow'),
    Supplier('Supplier2', 'SPb'),
    Supplier('Supplier3', 'Moscow')
]
print(suppliers_by_city(suppliers, 'Moscow'))

# е
from collections import OrderedDict
def keys_with_value_starting(od, start_letter):
    result = []
    for key, value in od.items():
        if str(value).lower().startswith(start_letter.lower()):
            result.append(key)
    return result
od = OrderedDict([('a', 'apple'), ('b', 'banana'), ('c', 'apricot'), ('d', 'orange')])
print(keys_with_value_starting(od, 'a'))
```

```python
# №14

# a
from collections import Counter
number = 123456123
digit_counts = Counter(str(number))
print(digit_counts)

# б
from collections import defaultdict
def countries_by_first_letter(countries):
    dd = defaultdict(list)
    for country, capital in countries:
        first_letter = country[0].upper()
        dd[first_letter].append((country, capital))
    return dict(dd)
countries = [('Russia', 'Moscow'), ('Germany', 'Berlin'), ('France', 'Paris'), ('Romania', 'Bucharest')]
print(countries_by_first_letter(countries))

# в
from enum import Enum
class TransportSpeed(Enum):
    CAR = 1
    BICYCLE = 2
    PLANE = 3
def max_transport_speed(transport):
    speeds = {
        TransportSpeed.CAR: 250,
        TransportSpeed.BICYCLE: 50,
        TransportSpeed.PLANE: 900
    }
    return speeds[transport]
print(max_transport_speed(TransportSpeed.PLANE))

# г
def unique_chars_from_strings(strings):
    chars = set()
    for s in strings:
        chars.update(s)
    return frozenset(chars)
strings = ['hello', 'world', 'python']
print(unique_chars_from_strings(strings))

# д
from collections import namedtuple
Client = namedtuple('Client', ['name', 'city', 'email'])
def clients_by_city(clients, target_city):
    return [client for client in clients if client.city == target_city]
clients = [
    Client('Alice', 'Moscow', 'alice@example.com'),
    Client('Bob', 'SPb', 'bob@example.com'),
    Client('Charlie', 'Moscow', 'charlie@example.com')
]
print(clients_by_city(clients, 'Moscow'))

# е
from collections import OrderedDict
def filter_by_key_start(od, start_letter):
    new_od = OrderedDict()
    for key, value in od.items():
        if key.lower().startswith(start_letter.lower()):
            new_od[key] = value
    return new_od
od = OrderedDict([('apple', 100), ('apricot', 120), ('banana', 50), ('orange', 80)])
filtered = filter_by_key_start(od, 'a')
print(filtered)
```

```python
# №15

# a
from collections import Counter
text = "hello world"
letter_counts = Counter(c for c in text.lower() if c.isalpha())
print(letter_counts)

# б
from collections import defaultdict
def first_three_letters_count(strings):
    dd = defaultdict(int)
    for s in strings:
        key = s[:3].upper()
        dd[key] += 1
    return dict(dd)
strings = ['apple', 'apricot', 'banana', 'application', 'band', 'cat']
print(first_three_letters_count(strings))

# в
from enum import Enum
class Gem(Enum):
    DIAMOND = 1
    EMERALD = 2
    RUBY = 3
def gem_hardness(gem):
    hardness = {
        Gem.DIAMOND: 10,
        Gem.EMERALD: 7.5,
        Gem.RUBY: 9
    }
    return hardness[gem]
print(gem_hardness(Gem.DIAMOND))

# г
def symmetric_difference_frozenset(set1, set2):
    return frozenset(set1 ^ set2)
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print(symmetric_difference_frozenset(set1, set2))

# д
from collections import namedtuple
Vehicle = namedtuple('Vehicle', ['brand', 'model', 'year'])
def vehicles_after_year(vehicles, year):
    return [v for v in vehicles if v.year > year]
vehicles = [
    Vehicle('Toyota', 'Camry', 2018),
    Vehicle('Honda', 'Accord', 2015),
    Vehicle('BMW', 'X5', 2020)
]
print(vehicles_after_year(vehicles, 2016))

# е
from collections import OrderedDict
def filter_by_substring(od, substring):
    new_od = OrderedDict()
    for key, value in od.items():
        if substring in str(value):
            new_od[key] = value
    return new_od
od = OrderedDict([('a', 'apple pie'), ('b', 'banana bread'), ('c', 'cherry cake'), ('d', 'apple juice')])
filtered = filter_by_substring(od, 'apple')
print(filtered)
```
