```python
# №1

import math

def midpoint(x1, y1, x2, y2):
    return (x1 + x2) / 2, (y1 + y2) / 2

def can_form_triangle(a, b, c):
    return a + b > c and a + c > b and b + c > a

def triangle_area_heron(a, b, c):
    if can_form_triangle(a, b, c):
        s = (a + b + c) / 2
        return math.sqrt(s * (s - a) * (s - b) * (s - c))
    else:
        return None

x1, y1 = 0, 0
x2, y2 = 4, 0
x3, y3 = 0, 3

seg_len1 = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
seg_len2 = math.sqrt((x3 - x1)**2 + (y3 - y1)**2)
seg_len3 = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)

print("Середина отрезка (1-2):", midpoint(x1, y1, x2, y2))
print("Можно ли построить треугольник?", can_form_triangle(seg_len1, seg_len2, seg_len3))
print("Площадь треугольника:", triangle_area_heron(seg_len1, seg_len2, seg_len3))
```

```python
# №2

import math

def distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

def triangle_perimeter(x1, y1, x2, y2, x3, y3):
    a = distance(x1, y1, x2, y2)
    b = distance(x2, y2, x3, y3)
    c = distance(x1, y1, x3, y3)
    if a + b > c and a + c > b and b + c > a:
        return a + b + c
    else:
        return None

def triangle_area(x1, y1, x2, y2, x3, y3):
    a = distance(x1, y1, x2, y2)
    b = distance(x2, y2, x3, y3)
    c = distance(x1, y1, x3, y3)
    if a + b > c and a + c > b and b + c > a:
        s = (a + b + c) / 2
        return math.sqrt(s * (s - a) * (s - b) * (s - c))
    else:
        return None

x1, y1 = 0, 0
x2, y2 = 4, 0
x3, y3 = 0, 3

print("Расстояния:", distance(x1, y1, x2, y2), distance(x2, y2, x3, y3), distance(x1, y1, x3, y3))
print("Периметр:", triangle_perimeter(x1, y1, x2, y2, x3, y3))
print("Площадь:", triangle_area(x1, y1, x2, y2, x3, y3))
```

```python
# №3

import math
import fractions

def gcd(a, b):
    return math.gcd(a, b)

def reduce_fraction(numerator, denominator):
    g = gcd(numerator, denominator)
    return numerator // g, denominator // g

def fraction_to_decimal(numerator, denominator):
    return numerator / denominator

num, den = 6, 8
print("НОД:", gcd(num, den))
print("Сокращенная дробь:", reduce_fraction(num, den))
print("Десятичная дробь:", fraction_to_decimal(num, den))
```

```python
# №4

def line_equation(x1, y1, x2, y2):
    if x2 - x1 == 0:
        return None, x1
    k = (y2 - y1) / (x2 - x1)
    b = y1 - k * x1
    return k, b

def are_lines_parallel(k1, b1, k2, b2):
    return k1 == k2

def is_point_on_line(k, b, x, y):
    return y == k * x + b

k1, b1 = line_equation(0, 0, 1, 2)
k2, b2 = line_equation(0, 1, 1, 3)
print("Уравнение первой прямой: y =", k1, "* x +", b1)
print("Прямые параллельны?", are_lines_parallel(k1, b1, k2, b2))
print("Точка (1, 2) на первой прямой?", is_point_on_line(k1, b1, 1, 2))
```

```python
# №5

import math

def circle_area(radius):
    return math.pi * radius**2

def circle_circumference(radius):
    return 2 * math.pi * radius

def sector_area(radius, angle_degrees):
    return (math.pi * radius**2) * (angle_degrees / 360)

def inscribed_square_perimeter(radius):
    side = radius * math.sqrt(2)
    return 4 * side

r = 5
angle = 90
print("Площадь круга:", circle_area(r))
print("Длина окружности:", circle_circumference(r))
print("Площадь сектора:", sector_area(r, angle))
print("Периметр вписанного квадрата:", inscribed_square_perimeter(r))
```

```python
# №6

def remove_char(string, char):
    return string.replace(char, "")

def replace_char(string, old, new):
    return string.replace(old, new)

def count_char(string, char):
    return string.count(char)

s = "hello world"
print("Удаление 'l':", remove_char(s, "l"))
print("Замена 'o' на 'a':", replace_char(s, "o", "a"))
print("Количество 'l':", count_char(s, "l"))
```

```python
# №7

def print_matrix(matrix):
    for row in matrix:
        print(row)

def matrix_add(A, B):
    if len(A) != len(B) or len(A[0]) != len(B[0]):
        return None
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def matrix_multiply(A, B):
    if len(A[0]) != len(B):
        return None
    result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result

def matrix_transpose(matrix):
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
print("Матрица A:")
print_matrix(A)
print("Сумма A+B:")
print_matrix(matrix_add(A, B))
print("Произведение A*B:")
print_matrix(matrix_multiply(A, B))
print("Транспонированная A:")
print_matrix(matrix_transpose(A))
```

```python
# №8

def vector_add(v1, v2):
    if len(v1) != len(v2):
        return None
    return [v1[i] + v2[i] for i in range(len(v1))]

def vector_scalar_multiply(v, scalar):
    return [x * scalar for x in v]

def vector_min_max(v):
    return min(v), max(v)

v1 = [1, 2, 3]
v2 = [4, 5, 6]
print("Сумма векторов:", vector_add(v1, v2))
print("Умножение на 2:", vector_scalar_multiply(v1, 2))
print("Минимум и максимум:", vector_min_max(v1))
```

```python
# №9

def remove_digits_from_file(filename):
    with open(filename, 'r') as f:
        content = f.read()
    new_content = ''.join([c for c in content if not c.isdigit()])
    with open(filename, 'w') as f:
        f.write(new_content)

def count_characters(filename):
    with open(filename, 'r') as f:
        return len(f.read())

def count_lines(filename):
    with open(filename, 'r') as f:
        return len(f.readlines())

filename = "test.txt"
with open(filename, 'w') as f:
    f.write("Hello123\nWorld456\n")
print("Исходный файл создан.")
print("Количество символов до:", count_characters(filename))
print("Количество строк до:", count_lines(filename))
remove_digits_from_file(filename)
print("Количество символов после:", count_characters(filename))
print("Количество строк после:", count_lines(filename))
```

```python
# №11

def fraction_power(num, den, power):
    return num**power, den**power

def fractions_equal(num1, den1, num2, den2):
    return num1 * den2 == num2 * den1

def fraction_greater(num1, den1, num2, den2):
    return num1 * den2 > num2 * den1

def fraction_less(num1, den1, num2, den2):
    return num1 * den2 < num2 * den1

num1, den1 = 1, 2
num2, den2 = 2, 4
print("Дробь 1 в степени 3:", fraction_power(num1, den1, 3))
print("Дроби равны?", fractions_equal(num1, den1, num2, den2))
print("Дробь 1 больше дроби 2?", fraction_greater(num1, den1, num2, den2))
```

```python
# №12

import math

def vector_subtract(v1, v2):
    return (v1[0] - v2[0], v1[1] - v2[1])

def dot_product(v1, v2):
    return v1[0]*v2[0] + v1[1]*v2[1]

def vector_length(v):
    return math.sqrt(v[0]**2 + v[1]**2)

v1 = (3, 4)
v2 = (1, 2)
print("Разность векторов:", vector_subtract(v1, v2))
print("Скалярное произведение:", dot_product(v1, v2))
print("Длина v1:", vector_length(v1))
```

```python
# №13

def insertion_sort(arr):
    arr = arr[:]
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def bubble_sort(arr):
    arr = arr[:]
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def selection_sort(arr):
    arr = arr[:]
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

arr = [64, 25, 12, 22, 11]
print("Исходный массив:", arr)
print("Сортировка вставками:", insertion_sort(arr))
print("Сортировка пузырьком:", bubble_sort(arr))
print("Сортировка выбором:", selection_sort(arr))
```

```python
# №14

def matrix_min(matrix):
    return min(min(row) for row in matrix)

def matrix_max(matrix):
    return max(max(row) for row in matrix)

def matrix_sum(matrix):
    return sum(sum(row) for row in matrix)

def row_sum(matrix, row_index):
    if 0 <= row_index < len(matrix):
        return sum(matrix[row_index])
    else:
        return None

mat = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print("Минимальный элемент:", matrix_min(mat))
print("Максимальный элемент:", matrix_max(mat))
print("Сумма всех элементов:", matrix_sum(mat))
print("Сумма строки 1:", row_sum(mat, 1))
```

```python
# №15

import math

def add_fractions(num1, den1, num2, den2):
    num = num1 * den2 + num2 * den1
    den = den1 * den2
    g = math.gcd(num, den)
    return num // g, den // g

def subtract_fractions(num1, den1, num2, den2):
    num = num1 * den2 - num2 * den1
    den = den1 * den2
    g = math.gcd(num, den)
    return num // g, den // g

def multiply_fractions(num1, den1, num2, den2):
    num = num1 * num2
    den = den1 * den2
    g = math.gcd(num, den)
    return num // g, den // g

num1, den1 = 1, 2
num2, den2 = 1, 3
print("Сумма:", add_fractions(num1, den1, num2, den2))
print("Разность:", subtract_fractions(num1, den1, num2, den2))
print("Произведение:", multiply_fractions(num1, den1, num2, den2))
```
